```python
df[["Age", "Credit amount", "Duration"]].hist(bins = 20, edgecolor = "black") # verifying numeric cols and displaying histogram
plt.suptitle("Distribution of Numerical Features", fontsize=14)
plt.show()
```

```python
# Creating box plots 
plt.figure(figsize=(10, 5))
for i, col in enumerate(["Age", "Credit amount", "Duration"]):
    plt.subplot(1, 3, i+1)
    sns.boxplot(y = df[col], color="skyblue")
    plt.title(col)

plt.tight_layout()
plt.show()
```
